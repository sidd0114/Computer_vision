```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# LOAD IMAGE
# -----------------------------
img = cv2.imread("spider man.jpg", 0)

if img is None:
    print("❌ Error: Image not loaded")
    exit()

# -----------------------------
# DEFINE FILTERS WITH NAMES
# -----------------------------
filters = [
    ("Vertical Edge Detection", np.array([[1, 0, -1],
                                          [1, 0, -1],
                                          [1, 0, -1]])),

    ("Horizontal Edge Detection", np.array([[1, 1, 1],
                                            [0, 0, 0],
                                            [-1, -1, -1]])),

    ("Sharpen Filter", np.array([[0, -1, 0],
                                 [-1, 5, -1],
                                 [0, -1, 0]]))
]

# -----------------------------
# Q1: INDEPENDENT FILTERING
# -----------------------------
plt.figure(figsize=(12,4))

plt.subplot(1, 4, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')

for i, (name, f) in enumerate(filters):
    out = cv2.filter2D(img, -1, f)

    plt.subplot(1, 4, i+2)
    plt.imshow(out, cmap='gray')
    plt.title(name)
    plt.axis('off')

plt.show()


# -----------------------------
# Q2: SEQUENTIAL FILTERING
# -----------------------------
seq_output = img.copy()

for name, f in filters:
    seq_output = cv2.filter2D(seq_output, -1, f)

    plt.imshow(seq_output, cmap='gray')
    plt.title(f"After {name}")
    plt.axis('off')
    plt.show()
```
